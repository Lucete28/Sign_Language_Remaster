```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, clear_output
import time
from tqdm import tqdm
# 넘파이 배열 생성 (예제용)
data = np.random.random((100, 30, 156))

# 반복 횟수
num_iterations = data.shape[0]

for i in tqdm(range(num_iterations)):
    # 넘파이 배열 선택
    d1 = data[i]

    # 시각화
    plt.imshow(d1, cmap='viridis')
    plt.colorbar()
    plt.title('Visualization of 30x156 Numpy Array')
    plt.show()

    # 출력 지우기
    time.sleep(1)
    clear_output(wait=True)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 예제용 100개의 30x156 배열 생성
# data2 = np.random.random((100, 30, 156))

# 이전 배열과의 유사도를 저장할 리스트
similarities = []
data2 = data
# 이전 배열 초기화
prev_array = data2[0].flatten()

# 반복문으로 각 배열과의 유사도 계산 및 시각화
for i in range(1, len(data2)):
    current_array = data2[i].flatten()

    # 코사인 유사도 계산
    cosine_similarity = np.dot(prev_array, current_array) / (np.linalg.norm(prev_array) * np.linalg.norm(current_array))

    # 유사도 저장
    similarities.append(cosine_similarity)

    # 히트맵으로 시각화

    # 현재 배열을 이전 배열로 설정
    prev_array = current_array

plt.plot(range(1, i+1), similarities, marker='o')
plt.title('Cosine Similarity with Previous Array')
plt.xlabel('Iteration')
plt.ylabel('Similarity')
plt.grid(True)
plt.show()
```
